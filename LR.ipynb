```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([[9], [12], [16], [19], [15]])  # Independent variable
Y = np.array([8, 12, 12.5, 14.9, 15])  # Dependent variable
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
Y_pred=model.predict(X)
```

```python
plt.scatter(X, Y, color='pink')  # Actual data points
plt.plot(X, Y_pred, color='blue')  # Regression line
plt.title("Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
print("Intercept (β0):", model.intercept_)
print("Coefficient (β1):", model.coef_)
```

```python
X = np.random.rand(50, 1) * 100  # 100 random numbers between 0 and 100
Y = 2 * X + np.random.randn(50, 1) * 10  # Linear relation with some noise
```
